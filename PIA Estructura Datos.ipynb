## **Método Groupby**


```python
## Se muestra el total de productos vendidos por ruta en el bimestre nov-dic-23
barra1=df.groupby('Ruta')['No. de productos vendidos'].sum()
barra1
```

```python
barra2= df.groupby('Ruta')['Devoluciones'].sum()
barra2
```

```python
ruta_tras1= df.groupby(['Ruta']).sum()
ruta_tras1
```

## **Gráfico de lineas**

```python
# Muestra el no. de productos vendidos y las devoluciones de cada ruta correspondientes al Bimestre Nov-Dic 2023

import matplotlib.pyplot as plt

plt.figure(figsize=(10,6))
plt.plot(barra1, label='No. de productos vendidos', color='blue', linewidth=4) # Grosor de la línea para los ingresos
plt.plot(barra2,label='Devoluciones', color='pink',linewidth=4)
plt.title('No. de productos vendidos y las Devoluciones por ruta GRUPO BIMBO Bimestre Nov-Dic 2023')
plt.xlabel('No. de productos vendidos')
plt.ylabel('Devoluciones')
plt.grid(True)
plt.legend()
plt.show()
```

## **Gráfico de pastel**

```python
#Muestra en porcentaje del total de ventas, el  no. de productos vendidos en cada Bimestre Nov-Dic

import matplotlib.pyplot as plt
colores = ['pink', 'yellow', 'blue', 'purple', 'red', 'green', 'orange', 'gray', 'brown', 'tan']
plt.figure(figsize=(12,8))
plt.pie(barra1, labels=barra1.index, colors=colores,  autopct='%1.1f%%')
plt.title('No. de productos vendidos en % de cada Ruta GRUPO BIMBO Bimestre Nov-Dic 2023')
plt.show()
```
